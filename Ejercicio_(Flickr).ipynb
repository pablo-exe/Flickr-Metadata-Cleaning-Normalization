# Limpieza de datos faltantes y repetidos

```python
#5a Obtenemos las columnas con datos faltantes en porcentaje

prueba = round(data.isnull().sum() / data.shape[0]* 100, 2)
prueba[prueba > 0]
```

```python
# 5b
# Eliminamos las columnas que tengan el 90% de los datos vacios

null_cols = data.isnull().sum()
drop = null_cols[null_cols > data.shape[0] * 0.9].index
data = data.drop(columns = list(drop))
```

```python
prueba = data.isnull().sum()
prueba[prueba > 0]
```

```python
# 5c Porcentaje de datos repetidos, .nunique nos da el numero de datos iguales en cada columna

(1 - data.nunique() / data.shape[0]) * 100
```

```python
# Procedemos a eliminar las columnas con datos repetidos superior al 90%

drop_cols = (1 - data.nunique() / data.shape[0]) * 100
drop_cols = drop_cols[(drop_cols > 90)].index
data = data.drop(drop_cols, axis=1)
data
```

```python
#Volvemos a ver la memoria que ocupa el archivo y vemos que hemos logrado reducirla

data.info(memory_usage = 'deep')
```

```python
# 6 Vamos a limpiar la columna de fechas
```

```python
import re

def search_pattern(value, pattern):
  search = re.search(pattern, str(value))
  if search:
    return search.group()
  elif str(value) == 'nan':
    pass
  else:
    print(value)
    return np.nan

# apply the custom function to the 'Date of Publication' column and match the pattern '\d{4}'
pattern = '\d{4}'
matches = data['Date of Publication'].apply(search_pattern, pattern=pattern)

# Update the dataframe
data['Date of Publication'] = matches
```

```python
data
```

```python
# Vamos a limpiar la columna de lugares
```

```python
pattern = '[a-zA-Z\s]+'

placeOfPublication = data['Place of Publication']
london_clean = placeOfPublication.str.contains('London|Londres')
ny_clean = placeOfPublication.str.contains('New York|Nueva York')
paris_clean = placeOfPublication.str.contains('Paris')

placeOfPublication.loc[london_clean] = 'London'
placeOfPublication.loc[ny_clean] = 'New York'
placeOfPublication.loc[paris_clean] = 'Paris'


matches = placeOfPublication.apply(search_pattern, pattern=pattern)

# Actualizamos el fichero original
data['Place of Publication'] = matches
```
